##### Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

##### Collatz function

```python
def collatz(valor, lista_x, lista_y):
    i = 0
    lista_x.append(i)
    lista_y.append(valor)
    while (valor != 1):  
        if (valor % 2 == 0):
            valor /= 2        
        else:
            valor = valor * 3 + 1
        i += 1
        lista_x.append(int(i))
        lista_y.append(int(valor))
    
    return (valor)
```

##### Mostrar ponto m√°ximo

```python
def ponto_maximo(x,y, ax=None):
    xmax = x[np.argmax(y)]
    ymax = np.max(y)
    text= "x={:0}, y={:0}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="-",connectionstyle="angle,angleA=0,angleB=120")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.96), **kw)
```

### Main

```python
valor = int(input("Digite o valor inicial: "))
lista_x = []
lista_y = []
```

```python

print ("Valor inicial: ", valor)
collatz(valor, lista_x, lista_y)

plt.plot(lista_x, lista_y, marker = 'o', ms = 3)
font1 = {'family':'arial', 'size':16}
ponto_maximo(lista_x, lista_y)
plt.title("Collatz conjecture", fontdict = font1)
plt.xlabel("Number of iterations", fontdict = font1)
plt.ylabel("Value", fontdict = font1)
plt.show()
```
